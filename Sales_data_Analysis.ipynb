# STEP 3: Load and View Data


```python
import pandas as pd

df = pd.read_csv('sales_data.csv')
df.head()
```

# STEP 4: Basic Data Understanding

```python
df.info()
df.describe()
```

# STEP 5: Total Revenue

```python
total_revenue = df['Revenue'].sum()
print("Total Revenue:", total_revenue)
```

# STEP 6: Sales by Product

```python
product_sales = df.groupby('Product')['Revenue'].sum().sort_values(ascending=False)
print(product_sales)
```

# STEP 7: Sales by Region

```python
region_sales = df.groupby('Region')['Revenue'].sum().sort_values(ascending=False)
print(region_sales)
```

# STEP 8: Visualizations

```python
import matplotlib.pyplot as plt

# Product Sales Chart
product_sales.plot(kind='bar')
plt.title('Revenue by Product')
plt.xlabel('Product')
plt.ylabel('Revenue')
plt.show()
```

```python
# Region Sales Chart
region_sales.plot(kind='bar')
plt.title('Revenue by Region')
plt.xlabel('Region')
plt.ylabel('Revenue')
plt.show()
```

```python
plt.figure(figsize=(8,5))
region_sales.plot(kind='bar')
plt.title('Revenue by Region')
plt.xlabel('Region')
plt.ylabel('Revenue')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
region_sales.plot(kind='bar')
plt.title('Revenue by Region')
plt.xlabel('Region')
plt.ylabel('Revenue')

plt.savefig('region_sales.png')
plt.show()
```

```python
plt.figure(figsize=(8,5))
product_sales.plot(kind='bar')

plt.title('Revenue by Product')
plt.xlabel('Product')
plt.ylabel('Revenue')
plt.xticks(rotation=0)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(8,5))

product_sales.plot(kind='bar', ax=ax)

ax.set_title('Revenue by Product')
ax.set_xlabel('Product')
ax.set_ylabel('Revenue')

plt.tight_layout()

fig.savefig('product_sales.png')  # This saves the correct figure
plt.show()
```
